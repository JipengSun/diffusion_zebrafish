```python
import tifffile
import matplotlib.pyplot as plt
import cv2
from skimage.util import random_noise
import numpy as np
```

```python
data_path = 'Data/volume100.tif'

tiff = tifffile.TiffFile(data_path)
# Iterate stack here
for page in tiff.pages:
    pass
print(len(tiff.pages))
# Access specific slice here
sample_image = tiff.pages[10]
    

print(sample_image.shape)
plt.figure(figsize=(15,15))
plt.imshow(sample_image.asarray())
```

```python
sample_image = tiff.pages[10]
sample_image_1 = sample_image.asarray()
gauss = np.random.normal(0,100,sample_image_1.size)
gauss = gauss.reshape(sample_image_1.shape[0],sample_image_1.shape[1]).astype('uint8')
# Add the Gaussian noise to the image
#img_gauss = cv2.add(sample_image_1,gauss)
sample_image_2 = sample_image_1 + gauss
plt.figure(figsize=(10,10))
plt.imshow(sample_image_2)
```
