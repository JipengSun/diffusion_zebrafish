```python
import tifffile
import matplotlib.pyplot as plt
import cv2
from skimage.util import random_noise
import numpy as np
import os
from pathlib import Path
from tifffile import imread, imwrite
from skimage.transform import resize
```

```python
def data_loader(paths, depth=14, resolution=128):
  output_path = "volume_dataset"

  for path in os.listdir(paths):
    print(path)
    # tiff = tifffile.TiffFile(paths+"/"+path)
    tiff = imread(paths+"/"+path)
    for i in range(depth):
      dst = f"{paths}/{output_path}/{i}"
      print(dst)
      Path(dst).mkdir(parents=True, exist_ok=True)
      # img = cv2.resize(img, dsize=(resolution, resolution), interpolation=cv2.INTER_CUBIC)
      img = tiff[i,:,:]
      resized_data = resize(img, (resolution, resolution), preserve_range=True, anti_aliasing=False)
      imwrite(f"{dst}/{path}", resized_data, planarconfig='CONTIG')
```
