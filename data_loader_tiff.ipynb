```python
import os
from tifffile import imread, imwrite
from pathlib import Path
path = "./Data/volume_dataset"
```

```python
def tiff2jpeg(paths):
  output_path = paths + '_jpeg'
  for path1 in os.listdir(paths):
    print(path1)
    if path1 != '.DS_Store':
      for path2 in os.listdir(f"{paths}/{path1}"):
        if path1 != 'DS_Store':
          path3 = f"{paths}/{path1}/{path2}"
          img = imread(path3)
          outfile = path2.split('.')[0] + '.jpg'
          dst = f"{output_path}/{path1}"
          # break
          Path(dst).mkdir(parents=True, exist_ok=True)
          print(f"{dst}/{outfile}")
          # break
          cv2.imwrite(f"{dst}/{outfile}", img, [int(cv2.IMWRITE_JPEG_QUALITY), 200])


tiff2jpeg(path)
```
