```python
import tifffile
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage,signal
```

```python
data_path = 'Data/Fish1_6.tiff'

tiff = tifffile.TiffFile(data_path)
for page in tiff.pages:
    pass
print(len(tiff.pages))
image_stack = tiff.pages[10]
    
#v = pims.TiffStack(data_path)
#v
print(image_stack.shape)
plt.figure(figsize=(15,15))
plt.imshow(image_stack.asarray())
```

```python

image_stack.dtype
a = ndimage.interpolation.zoom(image_stack,.117)
a = np.asarray(a)
print(a.shape)

MLA_dir = "Intermediate_Results/PSF_Stack/"
#np.save(MLA_dir+fname, psf_stack)
psf_stack = np.load(MLA_dir+'d1.8_f33.2_1200x1200_-40.0_40.0'+'.npy')
psf = np.abs(psf_stack[0])
plt.figure(figsize=(15,15))
plt.imshow(a)
plt.figure(figsize=(15,15))
plt.imshow(psf)
```
